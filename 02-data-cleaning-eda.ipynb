```python
# imports
import pandas as pd
import numpy as plt 
import matplotlib.pyplot as plt
import json
```

## Data Cleaning

```python
# Read in scraped data
model = pd.read_csv('data/model.csv')
engine = pd.read_csv('data/engine.csv')
body = pd.read_csv('data/body.csv')
trim = pd.read_csv('data/trim.csv')
```

```python
model
```

```python
# parsing json string to gather make: https://www.pluralsight.com/guides/convert-strings-to-json-objects-in-javascript-with-eval
model['make'] = model['make'].apply(lambda x: eval(x)['name'])
```

```python
model_drop = model.drop('make_id', axis=1)
model_drop
```

```python
engine
```

```python
# dropping unnecessary columns 
engine_drop = engine.drop(['make_model_trim_id', 'engine_type', 'fuel_type', 'cylinders',
                           'size', 'horsepower_hp', 'horsepower_rpm', 'torque_ft_lbs',
                           'torque_rpm', 'valves', 'valve_timing', 'cam_type', 'make_model_trim'],
                          axis=1)
engine_drop
```

```python
body
```

```python
body_drop = body.drop(['make_model_trim_id', 'length', 'width', 'height', 'wheel_base',
                       'front_track', 'rear_track', 'ground_clearance', 'cargo_capacity',
                       'max_cargo_capacity', 'curb_weight', 'gross_weight', 'max_payload',
                       'max_towing_capacity', 'make_model_trim'],
                      axis=1)
body_drop
```

```python
trim
```

```python
trim_drop = trim.drop(['make_model_id', 'name', 'description', 'invoice', 
                       'created', 'modified', 'make_model'],
                      axis=1)
trim_drop
```

## Joining the Data

```python
# Join Multiple Dataframes derived from 
# https://pandas.pydata.org/docs/getting_started/intro_tutorials/08_combine_dataframes.html

car = model_drop.merge(engine_drop, on='id').merge(body_drop, on='id').merge(trim_drop, on='id')
car
```

```python
# check for nulls
car.isnull().sum()
```

```python
# summary stats
car.describe()
```

```python
car.info()
```

```python
car['transmission'].value_counts()
```

```python
# replacing the transmission values into automatic vs manual
car['transmission'] = car['transmission'].replace(
    
    {
    '8-speed shiftable automatic': 'automatic',
    '10-speed shiftable automatic': 'automatic',
    '9-speed shiftable automatic': 'automatic',
    '7-speed automated manual': 'automatic',
    'continuously variable-speed automatic': 'automatic',
    '1-speed direct drive': 'automatic',
    '6-speed manual': 'manual',
    '8-speed automated manual': 'automatic',
    '6-speed shiftable automatic': 'automatic',
    '8-speed automatic': 'automatic',
    '7-speed shiftable automatic': 'automatic',
    '9-speed automatic': 'automatic',
    '5-speed manual': 'manual',
    '2-speed automated manual': 'automatic',
    '6-speed automated manual': 'automatic'
})
```

```python
car
```

```python
# export to final dataframe to csv
car.to_csv('data/car.csv', index=False)
```
