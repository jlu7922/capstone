```python
# imports
import pandas as pd
import numpy as plt 
import matplotlib.pyplot as plt

import requests
```

## Functions

```python
# formatting json data derived from this source: 'https://stackoverflow.com/questions/47577927'
# api 
def data_puller(starting_page, ending_page, url):
    ''' 
        Pulls .json data from API, from specified page range, with given url
        and formats it all into one dataframe.
    '''
    
    headers = {
        "X-RapidAPI-Key": "1201ae780cmsh7393c8d2372c07dp148e7fjsn029918f1d16c",
        "X-RapidAPI-Host": "car-api2.p.rapidapi.com"
    }

    data = []
    for page in range(starting_page, ending_page + 1):
        querystring = {"verbose":"yes","direction":"asc","sort":"id","page":page}
        response = requests.get(url, headers=headers, params=querystring)
        
        if response.status_code == 200:
            print(f"Connected to page {page} of {url}")
            page_data = response.json()['data']
            data.extend(page_data)
            
        else:
            print(f"Response status: {response.status_code}")
    
    return pd.DataFrame(data)
```

### pulling drivetrain and transmission types

```python
engine_url = "https://car-api2.p.rapidapi.com/api/engines"
data_puller(1, 10, engine_url)
```

### pulling body type

```python
body_url = "https://car-api2.p.rapidapi.com/api/bodies"
data_puller(1, 10, body_url)
```

### pulling car year / msrp data

```python
trim_url = "https://car-api2.p.rapidapi.com/api/trims"
data_puller(1, 10, trim_url)
```
